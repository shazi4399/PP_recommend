## 生成数据

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from sklearn import datasets

X, y = datasets.make_moons(n_samples=500, noise=0.3, random_state=666)
```

```python
plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.show()
```

```python
print(X)#X 是500 * 2维的数据
```

```python
print(y)# y 是 500 * 1维的label
```

```python
data_X = pd.DataFrame(X)
data_y = pd.DataFrame(y)
#拼接2张表
origin_data = pd.concat([data_X, data_y], axis=1)
```

```python
origin_data.columns = ['featureA','featureB', 'label']
```

```python
origin_data.head()
```

```python
origin_data.describe(include = 'all')
```

```python
# 判断X两维数据 和 y 的关系。
#draw a bar plot of survival by sex
sns.barplot(x = "featureA", y = "label", data = origin_data)
```
